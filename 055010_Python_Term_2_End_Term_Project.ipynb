- **Degrees of Freedom (df)**: 2

## Interpretation:
1. **Hypotheses**:
   - **Null Hypothesis (H₀)**: There is no association between "Shipping Method" and "Import_Export".
   - **Alternative Hypothesis (H₁)**: There is an association between "Shipping Method" and "Import_Export".

2. **Significance Level (α)**: 0.05
   - Since the p-value (0.024) is less than the significance level (0.05), we **reject the null hypothesis**.

3. **Conclusion**:
   - There is a **statistically significant association** between "Shipping Method" and "Import_Export". This suggests that the choice of shipping method depends on whether the shipment is an import or export.

4. **Comparison of Observed vs. Expected**:
   - The observed frequencies deviate slightly from the expected frequencies, contributing to the chi-squared statistic.

```python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row

# Calculate missing data summary for categorical subset
dh10_missing_summary_cat_col, dh10_missing_summary_cat_row = missing_data_summary(dh10_ds_cat)

# Calculate missing data summary for non-categorical subset
dh10_missing_summary_noncat_col, dh10_missing_summary_noncat_row = missing_data_summary(dh10_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
dh10_ds_cat_treated = dh10_ds_cat.dropna(thresh=len(dh10_ds_cat) * 0.5, axis=1)
dh10_ds_cat_treated = dh10_ds_cat_treated.dropna(thresh=len(dh10_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
dh10_ds_noncat_treated = dh10_ds_noncat.dropna(thresh=len(dh10_ds_noncat) * 0.5, axis=1)
dh10_ds_noncat_treated = dh10_ds_noncat_treated.dropna(thresh=len(dh10_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
dh10_dropped_variables_cat = len(dh10_ds_cat.columns) - len(dh10_ds_cat_treated.columns)
dh10_dropped_rows_cat = len(dh10_ds_cat) - len(dh10_ds_cat_treated)
dh10_dropped_variables_noncat = len(dh10_ds_noncat.columns) - len(dh10_ds_noncat_treated.columns)
dh10_dropped_rows_noncat = len(dh10_ds_noncat) - len(dh10_ds_noncat_treated)

dh10_total_dropped_variables = dh10_dropped_variables_cat + dh10_dropped_variables_noncat
dh10_total_dropped_rows = dh10_dropped_rows_cat + dh10_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(dh10_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(dh10_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(dh10_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(dh10_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {dh10_total_dropped_variables}")
print(f"Total rows dropped: {dh10_total_dropped_rows}")
```

```python
# Function to calculate descriptive statistics for categorical variables
def dh10_categorical_stats(df, cat_columns):
    dh10_results = []

    for dh10_col in cat_columns:
        dh10_counts = df[dh10_col].value_counts()
        dh10_proportions = dh10_counts / len(df)
        dh10_mode = dh10_counts.idxmax()
        dh10_min = df[dh10_col].min()
        dh10_max = df[dh10_col].max()

        for category, count in dh10_counts.items():
            dh10_results.append({
                "Column": dh10_col,
                "Category": category,
                "Frequency": count,
                "Proportion": dh10_proportions[category],
                "Minimum": dh10_min,
                "Maximum": dh10_max,
                "Mode": dh10_mode
            })

    return pd.DataFrame(dh10_results)

# Function to calculate correlation for categorical columns
def dh10_categorical_correlation(df, cat_columns, method='spearman'):
    dh10_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            dh10_col1, dh10_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[dh10_col1].astype('category').cat.codes,
                                        df[dh10_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[dh10_col1].astype('category').cat.codes,
                                         df[dh10_col2].astype('category').cat.codes)
            dh10_corr_results.loc[dh10_col1, dh10_col2] = corr
            dh10_corr_results.loc[dh10_col2, dh10_col1] = corr

    return dh10_corr_results

# Retrieve categorical columns from dh10_ds_cat_encoded
dh10_cat_columns = dh10_ds_cat_encoded.columns.tolist()

# Generate descriptive stats and correlations for these columns
dh10_categorical_stats_df = dh10_categorical_stats(dh10_ds_preprocessed, dh10_cat_columns)
dh10_spearman_corr_df = dh10_categorical_correlation(dh10_ds_preprocessed, dh10_cat_columns, method='spearman')
dh10_kendall_corr_df = dh10_categorical_correlation(dh10_ds_preprocessed, dh10_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
dh10_spearman_corr_df.fillna(1, inplace=True)
dh10_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(dh10_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(dh10_spearman_corr_df)

print("\nKendall Correlation:")
print(dh10_kendall_corr_df)
```

```python
# Descriptive Stats for Noncat Variables

# Function to calculate descriptive statistics for numerical variables
def dh10_numerical_stats(df, num_columns):
    dh10_results = []

    for dh10_col in num_columns:
        dh10_data = df[dh10_col].dropna()  # Exclude NaN values
        dh10_min = dh10_data.min()
        dh10_max = dh10_data.max()
        dh10_mean = dh10_data.mean()
        dh10_median = dh10_data.median()
        dh10_mode = dh10_data.mode().iloc[0] if not dh10_data.mode().empty else np.nan
        dh10_range = dh10_max - dh10_min
        dh10_std_dev = dh10_data.std()
        dh10_skewness = skew(dh10_data)
        dh10_kurt = kurtosis(dh10_data)

        # Coefficient of Variation (CV)
        dh10_cv = dh10_std_dev / dh10_mean if dh10_mean != 0 else np.nan

        # Confidence Interval (95%)
        dh10_conf_interval = norm.interval(0.95, loc=dh10_mean, scale=dh10_std_dev / np.sqrt(len(dh10_data)))

        # Add results
        dh10_results.append({
            "Column": dh10_col,
            "Minimum": dh10_min,
            "Maximum": dh10_max,
            "Mean": dh10_mean,
            "Median": dh10_median,
            "Mode": dh10_mode,
            "Range": dh10_range,
            "Standard Deviation": dh10_std_dev,
            "Skewness": dh10_skewness,
            "Kurtosis": dh10_kurt,
            "Coefficient of Variation": dh10_cv,
            "Confidence Interval (95%)": dh10_conf_interval
        })

    return pd.DataFrame(dh10_results)

# Function to calculate correlation for numerical variables
def dh10_numerical_correlation(df, num_columns, method='pearson'):
    dh10_corr_results = pd.DataFrame(index=num_columns, columns=num_columns, dtype=float)

    for i in range(len(num_columns)):
        for j in range(i, len(num_columns)):
            dh10_col1, dh10_col2 = num_columns[i], num_columns[j]
            if i == j:
                corr = 1.0  # Perfect correlation for the same variable
            else:
                if method == 'pearson':
                    corr, _ = pearsonr(df[dh10_col1].dropna(), df[dh10_col2].dropna())
                elif method == 'spearman':
                    corr, _ = spearmanr(df[dh10_col1].dropna(), df[dh10_col2].dropna())
            dh10_corr_results.loc[dh10_col1, dh10_col2] = corr
            dh10_corr_results.loc[dh10_col2, dh10_col1] = corr

    return dh10_corr_results

# Retrieve numerical columns from dh10_ds_noncat_scaled
dh10_num_columns = dh10_ds_noncat_scaled.columns.tolist()

# Generate statistics and correlations for these columns
dh10_numerical_stats_df = dh10_numerical_stats(dh10_ds_noncat_scaled, dh10_num_columns)
dh10_pearson_corr_df = dh10_numerical_correlation(dh10_ds_noncat_scaled, dh10_num_columns, method='pearson')
dh10_spearman_corr_df = dh10_numerical_correlation(dh10_ds_noncat_scaled, dh10_num_columns, method='spearman')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
dh10_pearson_corr_df.fillna(1, inplace=True)
dh10_spearman_corr_df.fillna(1, inplace=True)

# Display results
print("Numerical Statistics:")
print(dh10_numerical_stats_df)

print("\nPearson Correlation:")
print(dh10_pearson_corr_df)

print("\nSpearman Correlation:")
print(dh10_spearman_corr_df)
```

## Inferential Statistics

```python
# Select the two categorical variables for the test
dh10_categorical_var1 = 'Import_Export'
dh10_categorical_var2 = 'Shipping_Method'

# Create a contingency table
dh10_contingency_table = pd.crosstab(dh10_ds_preprocessed[dh10_categorical_var1],
                                     dh10_ds_preprocessed[dh10_categorical_var2])

# Print the contingency table
print("Contingency Table:")
print(dh10_contingency_table)

# Perform the Chi-squared test
dh10_chi2_stat, dh10_p_value, dh10_dof, dh10_expected_freq = chi2_contingency(dh10_contingency_table)

# Print the results
print(f"Chi-squared statistic: {dh10_chi2_stat}")
print(f"P-value: {dh10_p_value}")
print(f"Degrees of freedom: {dh10_dof}")
print("Expected frequencies:")
print(dh10_expected_freq)

# Interpret the results
dh10_alpha = 0.05  # Significance level
if dh10_p_value < dh10_alpha:
    print("Reject the null hypothesis. There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the variables.")
```

```python
# Assuming your preprocessed dataset is named 'dh10_ds_preprocessed'
# and your non-categorical columns are 'Quantity', 'Value', 'Weight'

# List of non-categorical variables
dh10_non_categorical_vars = ['Quantity', 'Value', 'Weight']

# Loop through each variable
for dh10_var in dh10_non_categorical_vars:
    print(f"\n--- Tests for {dh10_var} ---")

    # Shapiro-Wilk Test
    dh10_shapiro_stat, dh10_shapiro_p_value = stats.shapiro(dh10_ds_preprocessed[dh10_var])
    print(f"Shapiro-Wilk Test - Statistic: {dh10_shapiro_stat}, P-value: {dh10_shapiro_p_value}")
    if dh10_shapiro_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Kolmogorov-Smirnov Test
    dh10_ks_stat, dh10_ks_p_value = stats.kstest(dh10_ds_preprocessed[dh10_var], 'norm')
    print(f"Kolmogorov-Smirnov Test - Statistic: {dh10_ks_stat}, P-value: {dh10_ks_p_value}")
    if dh10_ks_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Anderson-Darling Test
    dh10_anderson_result = stats.anderson(dh10_ds_preprocessed[dh10_var])
    print("Anderson-Darling Test - Statistic:", dh10_anderson_result.statistic)
    # Check only for 5% significance level
    if dh10_anderson_result.statistic < dh10_anderson_result.critical_values[2]:  # Index 2 corresponds to 5%
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
    else:
        print(f"Data does not look normal (reject H0) at 5% significance level")

    # Jarque-Bera Test
    dh10_jb_stat, dh10_jb_p_value = stats.jarque_bera(dh10_ds_preprocessed[dh10_var])
    print(f"Jarque-Bera Test - Statistic: {dh10_jb_stat}, P-value: {dh10_jb_p_value}")
    if dh10_jb_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
```
